# Carry out correlation analysis on the dataset and find the correlation factor between a. ph and Chloramines b. ph and Conductivity What do we infer from this correlation value?

```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr
#a. ph and Chloramines
x=data['ph']
y=data['Chloramines']
# calculate Pearson's correlation
corr, _ = pearsonr(x, y)
print('Pearsons correlation: %.3f' % corr)
# calculate spearman's correlation
corr, _ = spearmanr(x, y)
print('Spearmans correlation: %.3f' % corr)
```

```python
#b. ph and Conductivity
x=data['ph']
y=data['Conductivity']
# calculate Pearson's correlation
corr, _ = pearsonr(x, y)
print('Pearsons correlation: %.3f' % corr)
# calculate spearman's correlation
corr, _ = spearmanr(x, y)
print('Spearmans correlation: %.3f' % corr)
```

In both cases correlation values are low.So in both cases variables have low positive correlation.
