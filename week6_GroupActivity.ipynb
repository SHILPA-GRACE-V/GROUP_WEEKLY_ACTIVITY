# 2. Examine the data by using visualizations

```python
#code by Daina maria
import matplotlib.pyplot as plt
import seaborn as sns
gdp=data[['country',"GDP per capita (current US$)"]]
top10=gdp.sort_values("GDP per capita (current US$)",ascending=False).head(10)
print(top10)
plt.figure(figsize=(12,8))
sns.barplot(x="country",y="GDP per capita (current US$)",data=top10)
plt.title("Top 10 countries based on GDP")
```

```python
# code by Daina maria
pop=data[['country',"Population in thousands (2017)"]]
top10=pop.sort_values("Population in thousands (2017)",ascending=False).head(10)
print(top10)
plt.figure(figsize=(12,8))
sns.barplot(x="country",y="Population in thousands (2017)",data=top10)
plt.title("Top 10 countries based on Population in thousands (2017)")
```

# 4. On average which region has the highest
## a. GDP per capita (current US)
##  b. International trade: Imports (million US$)

```python
#code by shilpa grace v
import numpy as np
#based on GDP
data.groupby("Region")["GDP per capita (current US$)"].mean().sort_values(ascending=False)
```

```python
import numpy as np
#based on International trade: Imports (million US$)
data.groupby("Region")["International trade: Imports (million US$)"].mean().sort_values(ascending=False)
```

NorthernAmerica has highest International trade: Imports (million US$)

# 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
#code by Wafaha k p
data1=data[['GDP per capita (current US$)','Region']].groupby("Region").std().sort_values(by=['GDP per capita (current US$)'],inplace=False)
data1
```

```python
D1=data.groupby("Region")["GDP per capita (current US$)"]
D2=D1.describe()
D2
```

```python
#Finding coefficient of variation
D3=D2['std']/D2['mean']*100
D3.sort_values()
```

Oceania have the lowest coefficient of variation therefor the most consistent
